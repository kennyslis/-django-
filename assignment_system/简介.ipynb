

# 📝 学生作业管理系统 - 功能与结构概览

## 📁 模型（models.py）

### CustomUser

自定义用户模型，继承 `AbstractUser`：

* `number`: 学号，唯一
* `name`: 姓名
* `email`: 邮箱
* `is_teacher`: 是否是老师
* `create_teacher`: 快速创建教师账户方法

### Assignment

布置的作业模型：

* `title`: 作业标题
* `description`: 描述
* `due_date`: 截止时间
* `file`: 附件

### Submission

学生提交模型：

* `student`: 提交人
* `assignment`: 所属作业
* `file`: 提交的`.ipynb`文件
* `submission_date`: 提交时间
* 自动重命名为 `学生名_作业名.ipynb`
* 提交后自动创建成绩

### Scores

记录每个学生每份作业的成绩：

* `student`: 学生
* `assignment`: 作业
* `score`: 成绩（默认0）
* `submission`: 对应提交（可为空）

## 🌐 视图功能（views.py）

### 学生功能：

* `/assignments/`: 学生查看所有作业，支持筛选未提交
* `/assignment/<id>/submit/`: 提交作业（限制 `.ipynb` 文件）
* `/check_grade`: 查看自己所有成绩
* `/update_profile`: 更新个人信息（含学号检查）
* `/forgot-password/`: 通过邮箱找回密码
* `/reset-password/`: 重设密码
* `/verify-code/`: 验证验证码后才可重设密码

### 老师功能：

* `/teacher/assignments/`: 作业管理面板（统计提交人数）
* `/teacher/assignments/create/`: 创建作业
* `/teacher/assignments/delete/<id>/`: 删除作业
* `/teacher/assigments/edit_assignment/<id>/`: 编辑作业
* `/assignment/<id>/submissions/`: 查看作业提交状态，支持筛选未提交
* `/assignment/<id>/submissions/<student_id>/grade/`: 对学生提交打分
* `/assignments/export/`: 导出所有成绩为 CSV
* `/assignment/<id>/export_non_submitted/`: 导出未提交名单
* `/change_password/`: 修改单个或所有学生密码
* `/teacher/grade_port/`: 查看作业提交状态（统一界面）
* `/teacher/change_pass/`: 密码管理界面（表单）

### 通用功能：

* `/import/`: Excel 导入学生信息（生成初始密码并发送邮件）
* `/login/`: 登录功能（教师跳转教师页）
* `/register/`: 注册功能（含是否为教师的标记）
* `/view-notebook/<submission_id>/`: 将 `.ipynb` 文件以 HTML 显示

---

## 📄 表单（forms.py）

* `SubmissionForm`：限制上传 `.ipynb` 文件
* `AssignmentForm`：用于创建/编辑作业，包含截止时间选择
* `ScoreForm`：老师打分界面使用

```python
path('assignments/', views.assignment_list, name='assignment_list')
path('assignment/<int:assignment_id>/submit/', views.student_submission, name='student_submission')
path('teacher/assignments/', views.teacher_assignment_management, name='teacher_assignment_management')
# ...
```
